```python
"""
Reverse list (without slicing)

👉 [1,2,3] → [3,2,1]
"""

inp = [1,2,3]
rev = []

for i in inp:
    rev.insert(0,i)

print(rev)
```

```python
"""
Q5: Palindrome check

👉 Input: "madam" → True
👉 Input: "hello" → False
"""
i1 = "madam"
ir1 =  i1[::-1]
print(True if ir1== i1 else False)

i2 = "hello"
ir2 = i2[::-1]
print(True if i2== ir2 else False)
```

```python
"""
Count vowels in string

👉 Input: "prajwal"
👉 Output: 2

👉 vowels = a, e, i, o, u

"""

vowels = ["a", "e", "i", "o", "u"]
inpu = "prajwal"

count = 0
for i in inpu:
    if i in vowels:
        count +=1
print(count)
```

```python
"""
Check if list is sorted

👉 Input: [1, 2, 3, 4] → True
👉 Input: [1, 3, 2] → False
"""

inp = [1,2,3,4]
is_sorted = True
pre = inp[0]
for i in inp:
    if i < pre:
        is_sorted = False
    pre = i
print(is_sorted)

inp0 = [1,3,2]
is_sorted0 = True
pre0 = inp0[0]

for i0 in inp0:
    if i0 < pre0:
        is_sorted0 = False
    pre0 = i0
print(is_sorted0)
```

```python
"""
Find duplicate elements

👉 Input: [1, 2, 2, 3, 3, 4]
👉 Output: [2, 3]

"""

inpu= [1, 2, 2, 3, 3, 4]

seen_elements = []
#print(s,"A")
duplicate = []
#print(dup,"B")
for i in inpu:
    if i in seen_elements:
        duplicate.append(i)
        #print(dup,"b")
        #print(s,"B")
    else:
        seen_elements.append(i)

print(duplicate)
```

```python
"""
🟡 Q3: Find second smallest element

👉 Input: [10, 5, 8, 20]
👉 Output: 8
"""
inp = [10, 8, 5, 20]

small = float('inf')
second = float('inf')

for i in inp:
    if i < small:
        second = small
        small = i
    elif i < second and i != small:
        second = i

print(second)
```

```python
"""
🔴 Q4: Move all zeros to end

👉 Input: [0, 1, 0, 3, 12]
👉 Output: [1, 3, 12, 0, 0]
"""

inp = [0, 1, 0, 3, 12]
zero = []
op = []
for i in inp:
    if i == 0:
        zero.append(i)
    else:
        op.append(i)
print(zero)
print(op)
for i in range(len(zero)):
    op.append(0)
print(op)
```

```python
"""
🔴 Q5: Find intersection of two lists

👉 Input:
[1, 2, 3, 4]
[3, 4, 5, 6]

👉 Output: [3, 4]
"""

l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]
op = []
for i in l1:
    for j in l2:
        if i == j:
            op.append(i)
print(op)
```
