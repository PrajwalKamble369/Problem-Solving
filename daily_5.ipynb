```python
"""
🔴 Q4: Find longest substring without repeating characters

👉 Input: "abcabcbb"
👉 Output: 3 (substring: "abc")
"""

inp = "abcabcbb"
max_len = 0

for i in range(len(inp)):
    seen = set()
    for j in range(i, len(inp)):
        if inp[j] in seen:
            break
        seen.add(inp[j])
    max_len = max(max_len, len(seen))

print(max_len)
```

```python
"""
🔴 Q4: Longest consecutive sequence

👉 Input: [100,4,200,1,3,2]
👉 Output: 4 (sequence: 1,2,3,4)
"""
inp = [100,4,200,1,3,2]
sorted_inp = sorted(inp)
first_element = sorted_inp[0]
second_element = sorted_inp[1]
diff = second_element - first_element
l1 = []
for i in range(len(inp)-1):
    fe = sorted_inp[i]
    se = sorted_inp[i+1]
    di = se - fe
    if di == diff:
        l1.append(sorted_inp[i])
print(l1)
    
```

```python
"""
🔴 Q5: Merge two sorted lists

👉 Input:
[1,3,5]
[2,4,6]

👉 Output: [1,2,3,4,5,6]
"""
l1 = [1,3,5]
l2 = [2,4,6]
l3 = {l1[0],}
for i in l1:
    for j in l2:
        if i >= j:
            l3.add(i)
        else:
            l3.add(j)

l4 = []
for i in l3:
    l4.append(i)
print(l4)
```
