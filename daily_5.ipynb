```python
"""
Check if list is sorted

👉 Input: [1, 2, 3, 4] → True
👉 Input: [1, 3, 2] → False
"""

inp = [1,2,3,4]
is_sorted = True
pre = inp[0]
for i in inp:
    if i < pre:
        is_sorted = False
    pre = i
print(is_sorted)

inp0 = [1,3,2]
is_sorted0 = True
pre0 = inp0[0]

for i0 in inp0:
    if i0 < pre0:
        is_sorted0 = False
    pre0 = i0
print(is_sorted0)
```

```python
"""
🟡 Q3: Find second smallest element

👉 Input: [10, 5, 8, 20]
👉 Output: 8
"""
inp = [10, 8, 5, 20]

small = float('inf')
second = float('inf')

for i in inp:
    if i < small:
        second = small
        small = i
    elif i < second and i != small:
        second = i

print(second)
```

```python
"""
🔴 Q4: Move all zeros to end

👉 Input: [0, 1, 0, 3, 12]
👉 Output: [1, 3, 12, 0, 0]
"""

inp = [0, 1, 0, 3, 12]
zero = []
op = []
for i in inp:
    if i == 0:
        zero.append(i)
    else:
        op.append(i)
print(zero)
print(op)
for i in range(len(zero)):
    op.append(0)
print(op)
```

```python
"""
🔴 Q5: Find intersection of two lists

👉 Input:
[1, 2, 3, 4]
[3, 4, 5, 6]

👉 Output: [3, 4]
"""

l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]
op = []
for i in l1:
    for j in l2:
        if i == j:
            op.append(i)
print(op)
```

```python
"""
🟡 Q3: Check if number is prime

👉 Input: 7 → True
👉 Input: 10 → False
"""

"""
Core Conditions
Natural Number: The number must be a positive integer ().
Divisors: The number must not be divisible by any integer other than 1 and itself.
Not 1: The number 1 is not considered a prime number because it has only one divisor.
Not 0 or Negative: 0 and negative numbers are not prime numbers
"""
inp = 10

is_prime = True

if inp <= 1:
    is_prime = False
else:
    for i in range(2, inp):
        if inp % i == 0:
            is_prime = False
            break

print(is_prime)
        
        
```

```python
"""
🟠 Q3: Find difference between max and min

👉 Input: [10,5,8,20]
👉 Output: 15
"""

inp = [10,5,8,20]
min_val = inp[0]
max_val = inp[0]
for i in inp:
    if i > max_val:
        max_val = i
    elif i < min_val:
        min_val = i

diff = max_val - min_val
print(diff)
```

```python
"""
🔴 Q5: Find all elements that appear only once

👉 Input: [1,2,2,3,4,4,5]
👉 Output: [1,3,5]
"""
inp = [1,2,2,3,4,4,5]
fer = {}
for i in inp:
    if i in fer:
        fer[i] += 1
    else:
        fer[i] = 1

un = []
for key,value in fer.items():
    if value == 1:
        un.append(key)
print(un)
```
