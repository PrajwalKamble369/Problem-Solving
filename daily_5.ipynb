```python
"""
Check if list is sorted

👉 Input: [1, 2, 3, 4] → True
👉 Input: [1, 3, 2] → False
"""

inp = [1,2,3,4]
is_sorted = True
pre = inp[0]
for i in inp:
    if i < pre:
        is_sorted = False
    pre = i
print(is_sorted)

inp0 = [1,3,2]
is_sorted0 = True
pre0 = inp0[0]

for i0 in inp0:
    if i0 < pre0:
        is_sorted0 = False
    pre0 = i0
print(is_sorted0)
```

```python
"""
🟡 Q3: Find second smallest element

👉 Input: [10, 5, 8, 20]
👉 Output: 8
"""
inp = [10, 8, 5, 20]

small = float('inf')
second = float('inf')

for i in inp:
    if i < small:
        second = small
        small = i
    elif i < second and i != small:
        second = i

print(second)
```

```python
"""
🔴 Q5: Find intersection of two lists

👉 Input:
[1, 2, 3, 4]
[3, 4, 5, 6]

👉 Output: [3, 4]
"""

l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]
op = []
for i in l1:
    for j in l2:
        if i == j:
            op.append(i)
print(op)
```

```python
"""
🟡 Q3: Check if number is prime

👉 Input: 7 → True
👉 Input: 10 → False
"""

"""
Core Conditions
Natural Number: The number must be a positive integer ().
Divisors: The number must not be divisible by any integer other than 1 and itself.
Not 1: The number 1 is not considered a prime number because it has only one divisor.
Not 0 or Negative: 0 and negative numbers are not prime numbers
"""
inp = 10

is_prime = True

if inp <= 1:
    is_prime = False
else:
    for i in range(2, inp):
        if inp % i == 0:
            is_prime = False
            break

print(is_prime)
        
        
```
