Mutable and Immutable Data Types

*Mutable object can changes its state or contents and immutable object cannot.*

1) Mutable Data Types
    - List
    - Dictionary
    - Byte Array
    - Set

2) Immutable Data Types
    - Int
    - Float
    - Complex
    - String
    - Tuple 
    

---- 

```python
print("""
Numbers:
      1) Integer
      2) Float
      3) Complex Number
\n""")

a = 5
b = 1.6
c = 3 + 9j
print("Integer is immutable data type,Integer Data Type:", type(a),"\n")
print("Float is immutable data type,,Float Data Type:", type(b),"\n")
print("Complex Number is immutable data type,Complex Number Data Type:",type(c),"\n")
```

```python
print("""
Sequence Data Type:
      1) String
      2) List
      3) Tupple
\n""")

a = "Prajwal"
b = [1,2,3,4]
c = (1,2,3,4)

print("String is immutable data type,String Data Type:",type(a),"\n")
print("List is mutable data type,List Data Type:",type(b),"\n")
print("Tuple is immutable data type,Tuple Data Type:",type(c),"\n")
```

```python
a = {"a":[1,2,3,4]}
b = {1,2,3,4}
print("Dictionary is mutable data type,Dictionary Data Type:",type(a),"\n")
print("Set is mutable data type,Set Data Type:",type(b),"\n")
```
